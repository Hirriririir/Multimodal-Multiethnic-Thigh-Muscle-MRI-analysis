```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Fatfraction = pd.read_excel('./Radiomics/Huashan fat fraction.xlsx')
Fatfraction
```

```python
Fatfraction_melted = Fatfraction.melt(id_vars='patient', var_name='Muscle', value_name='Fat_fraction')
Fatfraction_melted['patient'] = Fatfraction_melted['patient'] + '_' + Fatfraction_melted['Muscle']
Fatfraction_melted = Fatfraction_melted.drop(columns='Muscle')
Fatfraction_melted
```

```python
Radiomics = pd.read_excel("./Radiomics/Huashan_T1_radiomics.xlsx")
Radiomics = Radiomics.iloc[32:].T
Radiomics.columns = Radiomics.iloc[0]
Radiomics = Radiomics.iloc[1:]
Radiomics.reset_index(inplace=True)
Radiomics.rename(columns={'index': 'patient'}, inplace=True)
Radiomics
```

```python
merged_df = pd.merge(Fatfraction_melted, Radiomics, on='patient')
merged_df
```

```python


merged_df['Muscle'] = merged_df['patient'].str[-2:]

merged_df.to_excel('./Radiomics/Radiomics_fat_cor[groupby=muscle].xlsx', index=False)
```

```python
import scipy.stats  as stats


results = dict()
for column in df.columns[1:].to_list():
    correl = stats.pearsonr(df['Fat_fraction'], df[column])
    results[column] = correl

results = pd.DataFrame(results).T
results.columns = ['Correlation', 'P']
results.sort_values(by=['Correlation'], ascending=False)
```
